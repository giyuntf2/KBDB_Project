```python
from category_encoders import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler, MinMaxScaler, RobustScaler
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings('ignore')

#위에서 했던 상관계수로 순서분석
df = pd.read_csv('Train.csv')
df = df.drop(['ID'], axis=1)
df = df.rename(columns={'Reached.on.Time_Y.N':'ROT'})

target = ['ROT']
feature = df.drop(target, axis=1).columns

train, test = train_test_split(df, test_size=0.4, random_state=42)
train, val = train_test_split(train, test_size=0.4, random_state=42)
```

```python
X_train, y_train = train[feature], train[target]
X_val, y_val = val[feature], val[target]
X_test, y_test = test[feature], test[target]

X_train
```

```python
y_basemodel = y_train.copy()
y_basemodel.value_counts(normalize=True)
major = y_basemodel.mode().loc[0]
y_base_pred = [major] * len(y_basemodel)
accuracy_score(y_train, y_base_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer 
from sklearn.pipeline import make_pipeline
from scipy.stats import randint, uniform
from sklearn.model_selection import cross_val_score, RandomizedSearchCV, GridSearchCV
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.metrics import plot_confusion_matrix

#StandardSclaer
pipe_logis = make_pipeline(
    OneHotEncoder(),
    StandardScaler(),
    LogisticRegression(max_iter=1000),
)

pipe_logis.fit(X_train, y_train)
print('훈련 정확도', pipe_logis.score(X_train, y_train))
print('검증 정확도', pipe_logis.score(X_val, y_val))
print('Score함수 검증 정확도 : ', pipe_logis.score(X_val, y_val))
print("테스트세트 정확도 : ", pipe_logis.score(X_test, y_test))
  
fig, ax = plt.subplots(figsize=(5,5))
pcm = plot_confusion_matrix(pipe_logis, X_test, y_test, ax=ax) 
plt.show()
```

```python
#RobustScaler
pipe_logis2 = make_pipeline(
    OneHotEncoder(),
    RobustScaler(),
    LogisticRegression(),
)

pipe_logis2.fit(X_train, y_train)
print('훈련 정확도', pipe_logis2.score(X_train, y_train))
print('검증 정확도', pipe_logis2.score(X_val, y_val))

print('Score함수 검증 정확도 : ', pipe_logis2.score(X_val, y_val))
print("테스트세트 정확도 : ", pipe_logis2.score(X_test, y_test))
  
fig, ax = plt.subplots(figsize=(5,5))
pcm = plot_confusion_matrix(pipe_logis2, X_test, y_test, ax=ax) 
plt.show()
```

```python
#MMScaler
pipe_logis3 = make_pipeline(
    OneHotEncoder(),
    MinMaxScaler(),
    LogisticRegression(),
)

pipe_logis3.fit(X_train, y_train)
print('훈련 정확도', pipe_logis3.score(X_train, y_train))
print('검증 정확도', pipe_logis3.score(X_val, y_val))

print('Score함수 검증 정확도 : ', pipe_logis3.score(X_val, y_val))
print("테스트세트 정확도 : ", pipe_logis3.score(X_test, y_test))
  
fig, ax = plt.subplots(figsize=(5,5))
pcm = plot_confusion_matrix(pipe_logis3, X_test, y_test, ax=ax) 
plt.show()
```

```python
from ipywidgets import interact, fixed
from sklearn.metrics import classification_report

def explore_threshold(y_true, y_pred_proba, threshold=0.5):
    y_pred = y_pred_proba >= threshold
    vc = pd.Series(y_pred).value_counts()
    ax = sns.histplot(y_pred_proba, kde=True)
    ax.axvline(threshold, color='red')
    ax.set_title(f'# of target, 1={vc[1]}, 0={vc[0]}')
    plt.show()
    print(classification_report(y_true, y_pred))

    
interact(explore_threshold, 
    y_true=fixed(y_val), 
    y_pred_proba=fixed(y_pred_proba), 
    threshold=(0, 1, 0.01))
```
